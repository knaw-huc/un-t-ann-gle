```python
import json

with open('../data/1728/sessions-1728-output.json', 'r') as filehandle:
    json_data = json.loads(filehandle.read())
```

```python
for session in json_data['hits']['hits']:
    file_name = session['_id'] + '.json'
    with open('../data/1728/sessions-1728/'+file_name, 'w') as filehandle:
        json.dump(session, filehandle, indent=4)
```

```python
import requests
import json

def retrieve_res_json(query_string, date_string):
    response = requests.get(query_string)
    file_name = date_string + '-resolutions.json'
    
    with open('../data/1728/resolutions-1728/'+file_name, 'w') as filehandle:
        json.dump(response.json(), filehandle, indent=4)
```

```python
import glob
import re

path = "../data/1728/sessions-1728/meeting-1728-*"
res_query = 'https://annotation.republic-caf.diginfra.org/elasticsearch/pagexml_resolutions/_search?q=metadata.meeting_date:'

session_file_names = (f for f in glob.glob(path))
for n in session_file_names:
    date_string = re.search(r'meeting-(\S+)-session',n).group(1)
    retrieve_res_json(res_query+date_string, date_string)
```
