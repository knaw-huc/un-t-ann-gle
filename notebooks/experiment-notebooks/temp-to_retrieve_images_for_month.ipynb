Now filter on sessioin_date contains '1728-06'

```python
june_sessions = [s for s in sessions if '1728-06' in s['session_date']]
june_sessions
```

retrieve all full image IIIF urls

```python
iiif_urls = []
for s in june_sessions:   
    for img in s['image_range']:
        iiif_urls.append(img[0])
        
iiif_urls
```

for each jpg filename select the url  with the highest resolution. Note: each url corresponds with a column (i think)

```python
from urllib.parse import urlparse
```

```python
output = urlparse(iiif_urls[0])
output.path
```

```python
path_parts = output.path.split('/')
file_name = path_parts[4]
resolution = path_parts[6][1:]
print(f'file_name: {file_name}, resolution: {resolution}')
```

```python
url_records = []
for url in iiif_urls:
    path_parts = urlparse(url).path.split('/')
    url_record = {
        'url': url,
        'file_name': path_parts[4],
        'resolution': path_parts[6][1:]
    }
    url_records.append(url_record)
```

```python
from operator import itemgetter
from itertools import groupby

# iterate in groups
selected_iiif_url_recs = []

for file_name, items in groupby(url_records, key=itemgetter('file_name')):
    # first, convert the 'items' iterator to a list, to able to use it twice (iterators can be used once)
    itemlist = list(items)
    
    # copy the item with the lowest begin_index
    selected_iiif_url_recs.append(max(itemlist, key=itemgetter('resolution')).copy())
    
selected_iiif_url_recs
```

Final step: get these 50 jpg files for june 1728

```python
import requests

for rec in selected_iiif_url_recs:
    img_data = requests.get(rec['url']).content
    with open(rec['file_name'], 'wb') as handler:
        handler.write(img_data)
```
