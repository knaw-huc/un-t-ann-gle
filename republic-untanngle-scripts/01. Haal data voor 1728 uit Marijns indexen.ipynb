```python
import json

with open('../../data/1728/CAF-json/sessions-1728-output-18jul22.json', 'r') as filehandle:
    json_data = json.loads(filehandle.read())
```

```python
for session in json_data['hits']['hits']:
    file_name = session['_id'] + '.json'
    with open('../../data/1728/18jul22/CAF-sessions-1728-180722/'+file_name, 'w') as filehandle:
        json.dump(session, filehandle, indent=4)
```

```python
import requests
import json

def retrieve_res_json(query_string, date_string):
    response = requests.get(query_string)
    file_name = date_string + '-resolutions.json'
    
    with open('../../data/1728/18jul22/CAF-resolutions-1728-180722/'+file_name, 'w') as filehandle:
        json.dump(response.json(), filehandle, indent=4)
```

```python
import glob
import re
import os

path = "../../data/1728/18jul22/CAF-sessions-1728-180722/session-1728-*"
res_query = 'https://annotation.republic-caf.diginfra.org/elasticsearch/resolutions/_doc/_search?size=1000&track_total_hits=true&q=metadata.session_id:'

session_file_names = (f for f in glob.glob(path))
for n in session_file_names:
#    date_string = re.search(r'session-(\S+)-num',n).group(1)
    base = os.path.basename(n)
    session_id = os.path.splitext(base)[0]
    retrieve_res_json(res_query+f'"{session_id}"', session_id)
```
